```python
import cv2
import matplotlib.pyplot as plt
```

```python
# citra RGB
im_rgb = cv2.imread('image_rgb.jpg')
im_rgb = cv2.cvtColor(im_rgb, cv2.COLOR_BGR2RGB)

# Tampilkan citra asli RGB
plt.imshow(im_rgb)
plt.title('Citra RGB Asli')
plt.axis('off')
plt.show()
```

```python
# Baca citra RGB
im_gray = cv2.imread('image_gray.jpg')

# Ubah citra RGB ke citra grayscale
im_gray = cv2.cvtColor(im_gray, cv2.COLOR_BGR2GRAY)

# Tampilkan citra asli grayscale
plt.imshow(im_gray, cmap='gray')
plt.title('Citra Grayscale Asli')
plt.axis('off')
plt.show()
```

```python
# Menerapkan filter canny pada citra RGB
filter_canny_rgb = cv2.Canny(cv2.cvtColor(im_rgb, cv2.COLOR_RGB2GRAY), 25, 255, L2gradient=False)

# Tampilkan hasil deteksi tepi menggunakan metode Canny pada citra RGB
plt.imshow(filter_canny_rgb, cmap='gray')
plt.title('Deteksi Tepi dengan Metode Canny (RGB)')
plt.axis('off')
plt.show()
```

```python
# Menerapkan filter canny pada citra grayscale
filter_canny_gray = cv2.Canny(im_gray, 25, 255, L2gradient=False)

# Tampilkan hasil deteksi tepi menggunakan metode Canny pada citra grayscale
plt.imshow(filter_canny_gray, cmap='gray')
plt.title('Deteksi Tepi dengan Metode Canny (Grayscale)')
plt.axis('off')
plt.show()
```
